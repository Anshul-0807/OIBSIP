Convert Date into Day,Month,Year

```python
df['Date'] = pd.to_datetime(df[' Date'])
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month_name()
df['Year'] = df['Date'].dt.year
```

```python
df
```

### Analysis

YEAR AND MONTH WISE ANALYSIS:-

```python
year_month_average = df.groupby(['Year', 'Month'])[' Estimated Unemployment Rate (%)'].mean().reset_index()
```

```python
# Year_wise analysis

plt.figure(figsize=(10, 8))
plt.subplot(2, 1, 1)
plt.plot(year_month_average['Year'], year_month_average[' Estimated Unemployment Rate (%)'], marker='o', linestyle='-')
plt.title('Year_wise Analysis')
plt.xlabel('Year')
plt.ylabel('Average Unemployment Rate (%)')
plt.tight_layout()
plt.show()
```

```python
# month_wise analysis

plt.figure(figsize=(10, 8))
plt.subplot(2, 1, 2)
plt.plot(year_month_average['Month'], year_month_average[' Estimated Unemployment Rate (%)'], marker='o', linestyle='-')
plt.title('Month_wise Analysis')
plt.xlabel('Month')
plt.ylabel('Average Unemployment Rate (%)')
plt.tight_layout()
plt.show()
```

TIME-SERIES PLOT FOR UNEMPLOYEMENT RATE :-

```python
plt.figure(figsize=(10, 6))
df[' Estimated Unemployment Rate (%)'].plot(marker='o')
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend()
plt.grid(True)
plt.show()
```

HISTOGRAM OF UNEMPLOYEMENT RATE:

```python
plt.figure(figsize=(8, 6))
plt.hist(df[' Estimated Unemployment Rate (%)'], bins=20, edgecolor='k', color='crimson', alpha=0.8)
plt.title('Histogram of Unemployment Rate')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

IDENTIFY OUTLIERS THROUGH BOX-PLOT

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='Region', y=' Estimated Unemployment Rate (%)',
            data=df,
            width=0.6, 
            linewidth=1.5,  
            notch=True, 
            showfliers=False)
plt.title('Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Unemployment-Rate (%)')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

HEATMAP RELATION:-

```python
plt.figure(figsize=(12, 8))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='inferno', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

Bar-chart :-

```python
# bar chart of month unemployement average

monthly_avg = df.groupby(df['Date'].dt.strftime('%B'))[' Estimated Unemployment Rate (%)'].mean().sort_index()

plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_avg.index, y=monthly_avg.values, order=monthly_avg.index)
plt.title('Average Unemployment Rate by Month')
plt.xlabel('Month')
plt.ylabel('Average Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
